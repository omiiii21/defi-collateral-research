```python
from web3 import Web3
import json

infura_url = 'https://mainnet.infura.io/v3/a77f7fa2b1684b61b8de38395308ae51'
w3 = Web3(Web3.HTTPProvider(infura_url))
assert w3.is_connected(), "Failed to connect to Infura!"
```

```python
# default collateral factory contract details
factory_address = '0x1BC8FCFbE6Aa17e4A7610F51B888f34583D202Ec'  

import json
with open('defaultcollateralABI.json', 'r') as abi_file:
    defaultcollateralABI = json.load(abi_file)

factory_contract = w3.eth.contract(address=factory_address, abi=defaultcollateralABI)
```

```python
factory = w3.eth.contract(address=factory_address, abi=defaultcollateralABI)

# to get total number of deployed collateral contracts
total = factory.functions.totalEntities().call()
print(f"Total deployed DefaultCollateral contracts: {total}")

collateral_addresses = []
for i in range(total):
    addr = factory.functions.entity(i).call()
    collateral_addresses.append(addr)
    print(f"Entity {i}: {addr}")
```

```python

collateral_input_tokens = {}

for addr in collateral_addresses:
    collateral_contract = w3.eth.contract(address=addr, abi=defaultcollateralABI)
    
    try:
        input_token = collateral_contract.functions.asset().call()
    except Exception as e:
        print(f"Error fetching input token for {addr}: {e}")
        input_token = None
    
    collateral_input_tokens[addr] = input_token
    print(f"Collateral Contract: {addr} uses input token: {input_token}")

```

```python
import json
with open('ERC20-DAI.json', 'r') as abi_file:
    erc20_contract_abi = json.load(abi_file)
```

```python

factory_address = '0x1BC8FCFbE6Aa17e4A7610F51B888f34583D202Ec'
factory = w3.eth.contract(address=factory_address, abi=defaultcollateralABI)  

total = factory.functions.totalEntities().call()
print(f"Total deployed DefaultCollateral contracts: {total}")




for i in range(total):
    addr = factory.functions.entity(i).call()
    print(f"Entity {i}: {addr}")
    
    collateral_contract = w3.eth.contract(address=addr, abi=defaultcollateralABI)
    
    # Try to get the underlying asset address (if the function exists)
    try:
        asset_address = collateral_contract.functions.asset().call()
        print(f"  Underlying asset address: {asset_address}")
    except Exception as e:
        print(f"  Error retrieving asset for {addr}: {e}")
        continue  
    erc20_contract = w3.eth.contract(address=asset_address, abi=erc20_contract_abi)
    
    # Attempt to retrieve the token's name from the underlying asset
    try:
        token_name = erc20_contract.functions.name().call()
    except Exception as e:
        print(f"  Error retrieving token name for asset {asset_address}: {e}")
        token_name = "Unknown"
    
    print(f"  Underlying token name: {token_name}")
```
